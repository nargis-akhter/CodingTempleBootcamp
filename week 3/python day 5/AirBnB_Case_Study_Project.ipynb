```python
import pandas as pd
```

```python
ab_nyc_2019 = pd.read_csv(r'C:\Users\nargi\Documents\coding temple\week 3\python day 5\AB_NYC_2019.csv')
display(ab_nyc_2019)
```

```python
display(ab_nyc_2019.describe())
```

```python
ab_nyc_2019.dropna(subset = ['last_review'],inplace=True)
ab_nyc_2019.isna().sum()
```

```python
# Which hosts are the busiest and why?
busiest_host = ab_nyc_2019[ab_nyc_2019['availability_365'] < 1]
display(busiest_host)

display(busiest_host.isnull())
```

```python
# How many neighborhood groups are available and which shows up the most?
ab_nyc_2019.groupby(['neighbourhood_group']).count()

#Manhattan shows up the most
```

```python
ab_nyc_2019.groupby(['neighbourhood_group']).count().sort_values(['neighbourhood_group'], ascending=True)
```

```python
# Are private rooms the most popular in manhattan?
ab_nyc_2019.groupby(['neighbourhood_group', 'room_type']).count().sort_values(['neighbourhood_group'])['calculated_host_listings_count']

#No, Brooklyn has the most popular private rooms
```

```python
#Which neighorhood group has the highest average price?
display(ab_nyc_2019.groupby('neighbourhood_group').mean('price').round(2))

#Manhattan has the highest average price - $180.05 per night
```

```python
# Which neighbor hood group has the highest total price?
display(ab_nyc_2019.groupby('neighbourhood_group').max('price'))

#Brooklyn, Mahattan, and Queens
```

```python
#Which top 5 hosts have the highest total price?
ab_nyc_2019.sort_values('price', ascending=False).head(5)
```

```python
# Who currently has no (zero) availability with a review count of 100 or more?

zero_availability = busiest_host[busiest_host['number_of_reviews'] >= 100].sort_values('number_of_reviews', ascending=False)
new_stats = zero_availability[['host_name','number_of_reviews']].head(5)
display(new_stats)
new_stats.plot.area(x='host_name', y='number_of_reviews',stacked=False)

#Wanda from Brooklyn has zero availability with a review count of 480
```

```python
# What host has the highest total of prices and where are they located    

#Sonder (NYC) has the highest total of prices and they are located in the Financial District in Manhattan.
highest_prices_host_neighborhood = ab_nyc_2019.groupby(['host_id','host_name','neighbourhood','neighbourhood_group']).sum('price').round(0).sort_values('price', ascending=False).head(5).sort_values('host_id').reset_index()[[
    'host_name',
    'host_id',
    'host_name',
    'neighbourhood',
    'neighbourhood_group'
    ]
    ]
highest_prices_host_neighborhood
```

```python
# When did Danielle from Queens last receive a review?
daniel = ab_nyc_2019['host_name'] == 'Danielle'
queens = ab_nyc_2019['neighbourhood_group'] == 'Queens'
ab_nyc_2019[daniel & queens].sort_values('last_review', ascending=False)
```

## Further Questions

1. Which host has the most listings?

```python
ttl_host_listings = ab_nyc_2019.groupby(['host_id','host_name'],as_index=False).sum(
    'calculated_host_listings_count').sort_values('calculated_host_listings_count', ascending=False).head(5)[[
        'host_id','host_name','calculated_host_listings_count']]
ttl_host_listings

#Sonder (NYC) has the most listings at 67,689
```
