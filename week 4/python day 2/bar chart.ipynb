```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv("C:/Users/nargi/Downloads/bar_chart_data.csv")
df
```

```python
plt.figure(figsize = (9, 6))
plt.bar(x = df['Brand'], height = df['Cars Listings'], color = 'midnightblue')
plt.xticks(rotation = 45) #rotating labels is an easy way to utilize space without changing anything else in the plot
plt.title("Cars Listings by Brand")
```
